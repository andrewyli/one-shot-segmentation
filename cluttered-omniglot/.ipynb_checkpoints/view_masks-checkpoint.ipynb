```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

import os
```

```python
model_name = 'siamese-u-net'
LOG_DIR = os.path.join(os.getcwd(), 'logs/' + model_name + '/')
NUM_CHARS = 10
logdir = LOG_DIR + '%.d_characters/'%(NUM_CHARS)
```

```python
for index in range(10):
    im = np.load("images_sample_step_{}.npy".format(index))
    seg = np.load("labels_sample_step_{}.npy".format(index))
    tar = np.load("target_sample_step_{}.npy".format(index))

    plt.figure
    plt.subplot(131)
    plt.imshow(im)

    plt.subplot(132)
    plt.imshow(seg[...,0])

    plt.subplot(133)
    plt.imshow(tar)

    plt.show()
```

```python
vis_list = [5585, 6324, 1727, 1429, 142, 5490, 6568, 6676]

for index in vis_list:
    im = np.load(os.path.join(logdir, "vis", "sample_im_{}.npy").format(index))
    seg = np.load(os.path.join(logdir, "vis", "sample_seg_{}.npy").format(index))
    tar = np.load(os.path.join(logdir, "vis", "sample_tar_{}.npy").format(index))

    plt.figure
    plt.subplot(131)
    plt.imshow(im)

    plt.subplot(132)
    plt.imshow(seg[...,0])

    plt.subplot(133)
    plt.imshow(tar)

    plt.show()
```
