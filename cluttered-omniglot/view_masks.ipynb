```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
for index in range(10):
    im = np.load("images_sample_step_{}.npy".format(index))
    seg = np.load("labels_sample_step_{}.npy".format(index))
    tar = np.load("target_sample_step_{}.npy".format(index))

    plt.figure
    plt.subplot(131)
    plt.imshow(im)

    plt.subplot(132)
    plt.imshow(seg[...,0])

    plt.subplot(133)
    plt.imshow(tar)

    plt.show()
```
