```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

import os
```

```python
model_name = 'siamese-u-net'
LOG_DIR = os.path.join(os.getcwd(), 'logs/' + model_name + '/')
NUM_CHARS = 10
logdir = LOG_DIR + '%.d_characters/'%(NUM_CHARS)
```

```python
vis_list = [i for i in range(40)]

for index in vis_list:
    pred = np.load(os.path.join(logdir, "vis", "sample_pred_{}.npy").format(index))
    im = np.load(os.path.join(logdir, "vis", "sample_im_{}.npy").format(index))
    seg = np.load(os.path.join(logdir, "vis", "sample_gt_{}.npy").format(index))
    tar = np.load(os.path.join(logdir, "vis", "sample_tar_{}.npy").format(index))

    plt.figure(figsize=(30, 30))
    
    plt.subplot(141)
    plt.imshow(pred[...,0])
    
    plt.subplot(142)
    plt.imshow(im)

    plt.subplot(143)
    plt.imshow(seg[...,0])

    plt.subplot(144)
    plt.imshow(tar)

    plt.show()
```
